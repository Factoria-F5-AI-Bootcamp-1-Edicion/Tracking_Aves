# Leyendas de las amenazas de aves

#### Vamos a extraer una tabla con las leyendas de las amenazas para acompañar al mapa y que el usuario sepa que significan.

```python
#Importamos librerias
import pandas as pd
import numpy as np
```

```python
#Se lee pagina del catalogo de aves
leyen = pd.read_excel('./data_raw/Catalogacion_Especies_Aves_y_Planes_accion_12-2022.xlsx', sheet_name = 'Leyenda' )
leyen
```

```python
# Eliminamos las filas correspondientes a los títulos de cada comunidad, datos que no necesitamos.
leyen.drop([0, 1, 8, 14, 19, 23, 28, 33, 37, 41, 45, 49, 53, 58, 63, 66, 71, 76, 82, 87, 102, 103], axis=0, inplace=True)
leyen
```

#### Tenemos todas las leyendas, tanto las de amenaza como las de planes, vamos a separarlas.

```python
# Con la función loc partimos el dataframe, indicando de qué índice a qué índice.
leyen_amenaza = leyen.loc[:86]
leyen_planes = leyen.loc[88:]
```

```python
# Visualizamos para ver si quedó bien.
leyen_planes
```

```python
leyen_amenaza
```

```python
print(leyen_amenaza.columns.values)
```

```python
leyen_amenaza['D. Aves'].value_counts()
```

```python
leyen_amenaza.drop_duplicates(inplace=True)
leyen_amenaza['D. Aves'].value_counts()
```

```python
leyen_amenaza.drop([29, 75], axis=0, inplace=True)
```

```python
# Visualizamos, ya tenemos solo una vez cada leyenda con su significado.
leyen_amenaza
```

```python
# Cambiamos los nombres de las columnas
leyen_amenaza.rename(columns={'D. Aves' : 'Leyenda', 'Directiva Aves' : 'Significado'}, inplace=True)
leyen_amenaza.reset_index(inplace=True)
leyen_amenaza.drop(['index'], axis=1, inplace=True)
leyen_amenaza
```

#### Faltan algunas leyendas: NE, RE

```python
# Creamos los datos de las leyendas que faltan
leyendas_faltantes = pd.DataFrame({'Leyenda' : ['NE', 'RE'],
                                   'Significado' : ['No Evaluada', 'Regionalmente Extinta']},
                                   columns=['Leyenda', 'Significado'])
```

```python
leyen_amenaza = pd.concat([leyen_amenaza, leyendas_faltantes])
leyen_amenaza.reset_index(inplace=True)
leyen_amenaza.drop(['index'], axis=1, inplace=True)
leyen_amenaza
```

```python
# Cambiamos los nombres de las columnas de las leyendas de planes.
leyen_planes.rename(columns={'D. Aves' : 'Leyenda', 'Directiva Aves' : 'Significado'}, inplace=True)
leyen_planes.reset_index(inplace=True)
leyen_planes.drop(['index'], axis=1, inplace=True)
leyen_planes
```
