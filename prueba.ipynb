```python
import pandas as pd
import geopandas as gpd
```

```python
df = pd.read_csv("data_raw/Dataset_definitivo_con_geometrias.csv")
```

```python
df_puntos = df[df["NIVEL AMENAZA"]==1]
```

```python
aves_ciudades = df_puntos.groupby("Ubicacion")
aves_ciudades
```

```python
total_aves_ciudad=aves_ciudades.count()
```

```python
total_aves_ciudad
```

```python
print(total_aves_ciudad.columns.values)
```

```python
total_aves_ciudad.drop(['NOMBRE COMÚN','NOMBRE CIENTÍFICO','LR2004 (*Canarias)',
 'LISTA ROJA 2021 Invern./Migrat.',
 'LISTA ROJA 2021 Reproductoras (*Canarias)','CEEA y LESRPE', 'D. Aves',
 'AND', 'ARA', 'AST', 'BAL', 'CANA', 'CANT', 'CAT', 'CLM', 'CYL', 'EUSK', 'EXT',
 'GAL', 'MAD', 'MUR', 'NAV', 'RIO', 'VAL', 'EUROPEAN IUCN RED LIST BIRDS (2015)',
 'EUROPEAN IUCN RED LIST BIRDS (2021)',
 'GLOBAL UICN RED LIST (consulta 2022)',
 'GLOBAL UICN RED LIST (LAST ASSESSED)', 'CON_BERN', 'CON_BONN', 'CON_CITES',
 'Amenaza', 'index_ciudad', 'NIVEL AMENAZA'], axis=1,inplace=True) 
 
total_aves_ciudad
```

```python
total_aves_ciudad.reset_index(inplace=True) # resetear el indice, se convierte el indice por defecto
total_aves_ciudad
```

```python
print(total_aves_ciudad.columns.values)
```

```python
total_aves_ciudad.rename(columns={"Unnamed: 0":"TOTAL AVES"},inplace=True)
total_aves_ciudad
```

```python
mapapuntos
```

```python
print(mapapuntos.columns.values)
```

```python
mapapuntos.drop(['fecha_alta', 'fecha_baja', 'rotulo', 'id_ccaa', 'id_ine', 'nombres_f', 't_ua2'],axis=1,inplace=True)
mapapuntos
```

```python
total_aves_1= total_aves_ciudad.merge(mapapuntos,how="inner",left_on="Ubicacion",right_on="Ubicacion")
total_aves_1
```
